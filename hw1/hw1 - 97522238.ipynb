
In question, they asked us to open **background.png** in 3 channels mode.

if we use cv2.imread,in the default way, it loads images in 3 channels(however maybe, it was GRAYSCALE).
but in global we have 2 kind for loading an image :


1. CV_LOAD_IMAGE_GRAYSCALE => it means we want to load it with 2 channel
    

2. COLOR_BGR2GRAY => it means we want to load it with 3 channel (it is default value for loading)

so, for **loading background.png** in 3 channel way, we just need load it with cv2.imread()


### cv2.imread 

it gets 2 paramters :

  1. filename => Name of file to be loaded.


  2. flags =>	Flag that can take values of cv::ImreadModes (optional)


```python
import cv2
import os
from matplotlib import pyplot as plt

full_path = os.path.abspath(os.getcwd()) #get the full path of the current file directory

path = full_path+"/background.png"
img = cv2.imread(path)

# plt.imshow(img)
# plt.show()
```

## BGR and RGB
When reading a color image file, OpenCV imread() reads as a NumPy array ndarray of row (height) x column (width) x color (3). The order of color is BGR (blue, green, red).

The OpenCV function imwrite() that saves an image assumes that the order of colors is BGR, so it is saved as a correct image.

for changing it to RGB form, we just need to use this syntax :

    cv2.cvtColor(src,code)
    which our code is cv2.COLOR_BGR2RGB
    

```python
im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# plt.imshow(im_rgb)
# plt.show()
```

```python
background_path = full_path+"/background.png"
img_hold = cv2.imread(background_path)
line_thickness = 2
list_node = [(85, 145),(165, 145),(245, 145),(245, 145),(325, 145),(405, 145),(485, 145)]
main_node = [(5, 5),(5, 275),(565, 5),(565, 275)]

for node in main_node:
    for item in list_node :
        img_hold = cv2.line(img_hold, node, item, (255, 0, 255), thickness=line_thickness)
    print("adding {} node".format(str(node)))
    plt.imshow(img_hold)
    plt.show()

# cv2.imshow(img_end) 
img_end = cv2.imwrite("mypic.jpg", img_hold)

```

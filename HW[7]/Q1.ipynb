# Panorama with OpenCV

```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
from os import listdir
from os.path import isfile, join
```

```python
# read input victoria images and show them in a row together


main_path = "images/Q1"
onlyfiles = [f for f in listdir(main_path) if isfile(join(main_path, f))]

list_img = []
for img in onlyfiles:
    list_img.append(cv2.imread(main_path+"/"+img))



fig, (ax1, ax2, ax3, ax4, ax5, ax6, ax7) = plt.subplots(nrows=1, ncols=7, figsize=(14, 14),
                                    sharex=True, sharey=True)
for aa in (ax1, ax2, ax3, ax4, ax5, ax6, ax7):
    aa.set_axis_off()

ax1.imshow(list_img[0])
ax2.imshow(list_img[1])
ax3.imshow(list_img[2])
ax4.imshow(list_img[3])
ax5.imshow(list_img[4])
ax6.imshow(list_img[5])
ax7.imshow(list_img[6])


plt.tight_layout()
plt.show()
```

```python
# initialize OpenCV's image sticher object and then perform the image stitching on input images

dummy = not cv2.STITCHER_OK
while(dummy != cv2.STITCHER_OK):
    stitchy=cv2.Stitcher.create()
    (dummy,output)=stitchy.stitch(list_img)
```

```python
# show victoria panorama
plt.imshow(output)
```

```python
import cv2
import pytesseract
from google.colab.patches import cv2_imshow
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import required packages


##################### your code ################################
img = cv2.imread("/content/royan.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray,(5,5),0)


ret, thresh1 = cv2.threshold(blur, 1, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)     
# plt.imshow(thresh1)         

kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
dilation = cv2.dilate(thresh1,kernel,iterations = 1)

################################################################
# Finding contours
print('THRESH_OTSU: ')
print(pytesseract.image_to_string( dilation , lang='fas'))
```

```python
img = cv2.imread("/content/royan.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

th = cv2.adaptiveThreshold(gray,255,cv2.MORPH_RECT,\
            cv2.THRESH_BINARY,11,2)
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
dilation = cv2.dilate(th,kernel,iterations = 1)
print("ADAPTIVE_THRESH_GAUSSIAN_C :")
print(pytesseract.image_to_string( dilation , lang='fas'))
```

```python
# Import required packages
import cv2
import pytesseract
from google.colab.patches import cv2_imshow
from PIL import Image
##################### your code ################################
img = cv2.imread("/content/mypic.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray,(3,3),0)


ret, thresh1 = cv2.threshold(blur, 1, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)     
# plt.imshow(thresh1)         

kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
dilation = cv2.dilate(thresh1,kernel,iterations = 1)


################################################################
# Finding contours
im2 = img.copy()
contours, hierarchy = cv2.findContours(dilation, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)

for cnt in contours:
  x, y, w, h = cv2.boundingRect(cnt)
	# Drawing a rectangle on copied image
  rect = cv2.rectangle(im2, (x, y), (x + w, y + h), (0, 255, 0), 2)
	# Cropping the text block for giving input to OCR
  cropped = im2[y:y + h, x:x + w]
  print(pytesseract.image_to_string(cropped, lang='fas'))
```

```python
image_path_in_colab="/content/mypic.png"
extractedInformation = pytesseract.image_to_string(image_path_in_colab, lang='fas')
print(extractedInformation)
```
